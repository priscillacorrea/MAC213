## Analise de distribuicao de Sexo por atividadde

No data frame anterior, importamos o csv em sua totalidade. Se quisermos importar algumas colunas especificas, podemos utilizar
o argumento 'usecols'. Assim, podemos definir quais linhas queremos utilizar.
Vamos fazer uma analise da distribuicao dos sexos entre as atividades realizadas nas instituicoes.

```python
df = pd.read_csv('idadeserieneeracadez17.csv' , 
        sep = ";", encoding = 'latin-1', header = 0, usecols = ['DESCSERIE', 'SEXO' , 'QTDE'])
df
```

Assim, importamos apenas os dados relevantes para a analise que desejamos fazer.
Com o Comando axes, o pandas ira nos indicar algumas informacoes, como, quantas linhas tem nosso dataFrame, neste caso, 346861 e quais sao as colunas que importamos, que neste caso sao DESCSERIE, que indica a descricao da atividade, SEXO, que indica se o grupo e feminino ou masculino e QTDE que indica a quantidade de alunos do grupo respectivo.

```python
df.axes
```

## Funcao sum

Utilizando a funcao sum do panda Data frame, conseguimos somar todos os valores. 
Vamos supor que desejamos descobrir quantos alunos participam da atividade "CLUBE MATEMATIC".

Para isso, vamos criar uma variavel chamada mathClub.
Vamos atribuir a ela a soma das QTDE, de todas as linhas que tiveram "CLUBE MATEMATIC" como descricao da serie.
Para isso, escrevemos:


```python
mathClub = df[df["DESCSERIE"]=='CLUBE MATEMATIC'].sum()["QTDE"]
mathClub
```

Assim, conseguimos obter o valor total de matriculados em "CLUBE MATEMATIC". Se, no codigo acima, trocarmos "CLUBE MATEMATIC" por outro argumento, seremos capazes de descobrir as quantidades de qualquer outra atividade que conste em nosso csv.

## Criando uma pivot table

Vamos agora usar a funcao [pivot_table](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.pivot_table.html) para criar uma tabela.

Vamos criar uma tabela que, agrupa todas as quantidades de meninas(F) e meninos(M), indexa pela descricao da atividade(DESCSERIE).

Para tal, iremos criar uma variavel pivotSexo e chamar pivot_table sobre nosso dataframe df.





```python
pivotSexo = df.pivot_table(index = 'DESCSERIE', columns = 'SEXO', aggfunc=sum)
print(pivotSexo)
```

Assim, obtivemos uma tabela com as atividades ordenadas em ordem alfabeticas e com duas colunas, que indicam as quantidades de pessoas do sexo feminino e masculino que participam de tal atividade.

Vamos supor agora, que desejamos obter informacoes apenas sobre uma atividade. Por exemplo, CLUBE MATEMATIC. podemos usar o [loc](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.loc.html) para acessar a informacao que desejamos:

```python
mathClubSexo = pivotSexo.loc['CLUBE MATEMATIC']
print(mathClubSexo)
```

```python
violin = pivotSexo.loc['VIOLINO/VIOLA P']
print(violin['QTDE'])
```

Repare ainda, que, apesar de estarmos acessando informacoes contidas na pivot table que criamos, ela ainda esta intacta e podemos acessa-la novamente, e ela continua contendo as mesmas informacoes de antes:

```python
print(pivotSexo)
```

### MAX e IDMAX

Suponha agora que desejamos descobrir quais atividades tem a maior diferenca entre meninos e meninas inscritos?
Temos duas funcoes muito uteis, [idmax](https://pandas.pydata.org/pandas-docs/version/0.17.0/generated/pandas.DataFrame.idxmax.html) e [max](https://pandas.pydata.org/pandas-docs/version/0.17.0/generated/pandas.DataFrame.max.html).

IDMAX retorna o indice com maior valor. No nosso caso, iremos usa-la para retornar o nome da atividade com o maior valor definido. 

Ja MAX retorna o valor propriamente dito.

```python
print("Atividade com maior diferenca absoluta entre meninos e meninas : ")
print((pivotSexo['QTDE']['M'] - pivotSexo['QTDE']['F']).idxmax())

print("\nAtividade com maior diferenca absoluta entre meninas e meninos : ")
print((pivotSexo['QTDE']['F'] - pivotSexo['QTDE']['M']).idxmax())

print("\nAtividade com maior diferenca relativa entre meninos e meninas : ")
print(((pivotSexo['QTDE']['M'] - pivotSexo['QTDE']['F'])/(pivotSexo['QTDE']['M']+pivotSexo['QTDE']['F'])).idxmax())
print("Razao de meninos para meninas : ")
print(((pivotSexo['QTDE']['M'] - pivotSexo['QTDE']['F'])/(pivotSexo['QTDE']['M']+pivotSexo['QTDE']['F'])).max())

print("\nAtividade com maior diferenca relativa entre meninas e meninos : ")
print(((pivotSexo['QTDE']['F'] - pivotSexo['QTDE']['M'])/(pivotSexo['QTDE']['M']+pivotSexo['QTDE']['F'])).idxmax())
```
