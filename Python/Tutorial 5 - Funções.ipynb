# Funções 

Uma função é um bloco de código que apenas roda quando é chamada. Já usamos uma função, que vem definida por padrão no Python, o print. Mas podemos também criar nossas próprias funções, para realizarmos tarefas complexas, sem ter que reescrever muitas vezes o mesmo trecho de código.

```python
#Exemplo de funcao

def imprime(valor):
    print("valor digitado: " + str(valor))
    print("Dobro do valor digitado: " + str(valor*2))

imprime(5)
```

Vamos importar a base de dados [sobre Rendimento e movimento escolar](http://dados.prefeitura.sp.gov.br/dataset/rendimento-e-movimento-escolar-ensino-fundamental) e verificar dados sobre alunos aprovados, repovados e evadidos no ensino fundamental:


```python
import pandas as pd


#Lista de colunas que desejamos importar
cols = ["ANO", "TOTALEF_ABAND", "TOTALEF_APROV", "TOTALEF_REPROV"]

df = pd.read_csv('CSV/censoevolutivomovimentorendimento.csv' , 
                 sep = ";", encoding = 'latin-1', header = 0, usecols=cols)

#Utilizar apenas linhas cujo valor da coluna ano e 2015
df = df[df.ANO == 2015]

df['TOTAL'] = df['TOTALEF_ABAND'] + df['TOTALEF_APROV'] + df['TOTALEF_REPROV'] 

#Utilizar apenas linhas cujo valor da coluna Total e diferente de 0
df = df[df.TOTAL != 0].reset_index(drop=True)

df
```

Vamos agora definir uma função chamada razão, que ira retorna a razão entre dois ou mais valores:

```python
def razao(valor1, valor2):
    return valor1/valor2
```

Vamos criar uma nova coluna Razao_Aprov que mostra o total de alunos aprovados

```python
df['RAZAO_APROV'] = razao(df['TOTALEF_APROV'], df['TOTAL']) 

df
```

Vamos agora obter a média entre todas as instituições:

```python
media_Razao_Aprovados = df['RAZAO_APROV'].mean()
media_Razao_Aprovados
```

Dessa forma, conseguimos descobrir que, em média, 95.8% dos alunos foi aprovado no ensino fundamental como um todo, no ano de 2015.

Podemos ainda descobrir, entre os reprovados, qual porcentagem evadiu e quantos foram reprovados:

```python
df['TOTAL_NAPROV'] = df['TOTALEF_ABAND'] + df['TOTALEF_REPROV'] 
df['RAZAO_REPROV'] = razao(df['TOTALEF_REPROV'], df['TOTAL_NAPROV']) 
df['RAZAO_ABAND'] = razao(df['TOTALEF_ABAND'], df['TOTAL_NAPROV']) 

df
```

```python
media_Razao_Reprovados = df['RAZAO_REPROV'].mean()
media_Razao_Reprovados
```

```python
media_Razao_Aband = df['RAZAO_ABAND'].mean()
media_Razao_Aband
```

Vamos agora criar alguns gráficos para representar os valores que obtivemos:


```python
import matplotlib.pyplot as plt
%matplotlib inline  


fig = plt.pie([media_Razao_Aprovados, 1- media_Razao_Aprovados], labels = ['Aprovados', 'Não Aprovados'], 
              autopct='%1.f%%', startangle=45, explode = [0.0, 0.05])
plt.axis('equal') #Fazer o grafico ser um circulo perfeito
plt.title('Aprovados e Não Aprovados', fontsize = 'x-large') #adicionando o 
patches, texts, autotexts = fig
plt.setp(autotexts, size='medium', color = 'w', weight = 'bold')
plt.savefig('Exemplos/Aprovados.png', dpi = 300)
plt.show()
```

```python
fig = plt.pie([media_Razao_Reprovados, media_Razao_Aband], labels = ['Reprovados', 'Abandonos'], 
              autopct='%1.f%%', startangle=45, explode = [0.0, 0.05])
plt.axis('equal') #Fazer o grafico ser um circulo perfeito
plt.title('Não Aprovados', fontsize = 'x-large') #adicionando o 
patches, texts, autotexts = fig
plt.setp(autotexts, size='x-large', color = 'w', weight = 'bold')
plt.savefig('Exemplos/NAaprovados.png', dpi = 300)
plt.show()
```

Assim, conseguimos ver que 4% dos alunos não foi aprovado no ensino fundamental em 2015 e, desses, 87% foi reprovado e 13% abandonou a escola.
